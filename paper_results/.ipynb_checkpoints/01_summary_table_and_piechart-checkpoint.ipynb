# load regulamentary output

```python
re = pd.read_csv('metadata/huvec_mlv_REgulamentary.csv', sep='\t')
re
```

```python
new_re = []
for r, me1, me3, ctcf in zip(re['RE'], re['H3K4me1_rank'], re['H3K4me3_rank'], re['CTCF_rank']):
    
    if r=='TBD_1':
        if me1 <= me3:
            new_re.append("Promoter")
        else:
            new_re.append("Enhancer")

    elif r=='TBD_2':
        if me1 <= me3:
            new_re.append("Promoter/CTCF")
        else:
            new_re.append("Enhancer/CTCF")
            
    elif r=='TBD_3':
        if me1 <= me3:
            new_re.append("Promoter/CTCF")
        else:
            new_re.append("Enhancer/CTCF")
            
    else:
        new_re.append(r)
```

```python
re['new_re'] = new_re
```

```python
re = re[['chromosome', 'start', 'end', 'new_re']]
```

```python
results_lead = {}
b = pybedtools.BedTool.from_dataframe(lead)
for r in sorted(re['new_re'].value_counts().index):
    re_sub = re[re['new_re']==r]
    a = pybedtools.BedTool.from_dataframe(re_sub)
    ab = a.intersect(b).to_dataframe()
    results_lead[r] = ab
```

```python
results_imputed = {}
b = pybedtools.BedTool.from_dataframe(imputed)
for r in sorted(re['new_re'].value_counts().index):
    re_sub = re[re['new_re']==r]
    a = pybedtools.BedTool.from_dataframe(re_sub)
    ab = a.intersect(b).to_dataframe()
    results_imputed[r] = ab
```

# Pie Charts

```python
data_lead   = []
labels_lead = []

for k in results_lead.keys():
    if results_lead[k].shape[0]!=0:
        data_lead.append(results_lead[k].shape[0])
        labels_lead.append(k)
    
data_imputed   = []
labels_imputed = []

for k in results_imputed.keys():
    if results_imputed[k].shape[0]!=0:
        data_imputed.append(results_imputed[k].shape[0])
        labels_imputed.append(k)
```

```python
def make_autopct(values):
    def my_autopct(pct):
        total = sum(values)
        val = int(round(pct*total/100.0))
        return '{p:.2f}% ({v:d})'.format(p=pct,v=val)
    return my_autopct

colors = sns.color_palette('bright')

fig = plt.figure(figsize=(16,9), dpi=100)

ax1 = plt.subplot2grid((1,2),(0,0))
plt.pie(data_lead,    labels = labels_lead,    colors = colors, autopct=make_autopct(data_lead))
plt.title('Lead distribution')

ax1 = plt.subplot2grid((1, 2), (0, 1))
plt.pie(data_imputed, labels = labels_imputed, colors = colors, autopct=make_autopct(data_imputed))
plt.title('Imputed distribution')

fig.show()
```
